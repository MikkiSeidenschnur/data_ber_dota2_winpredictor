# Merge Data

**1. Find all files on rawdata folder and add it to a list**

```python
csv_files = [f for f in listdir("../rawdata/")]
```

**2. Use pandas to merge csv files**

```python
data = pd.DataFrame(columns=["match_id","match_seq_num", "radiant_win","start_time","duration","avg_mmr","num_mmr","lobby_type","game_mode","avg_rank_tier","num_rank_tier","cluster","radiant_team","dire_team"])
```

```python
for file in csv_files:
    new = pd.read_csv(f"../rawdata/{file}")
    data = data.append(new, ignore_index=True)
```
