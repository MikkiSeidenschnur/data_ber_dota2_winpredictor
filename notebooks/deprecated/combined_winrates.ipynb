```python
import pandas as pd
import numpy as np
```

```python
### Import dataset for winrate 
winrate_dataframe = pd.read_csv("../data/heroes_matchups.csv")
### Import dataset for test
test_set = pd.read_csv("../data/random_matches.csv")
winrate_dataframe.loc[(winrate_dataframe["hero"] == 69) & (winrate_dataframe["opponent"] == 105)]
```

```python
test_set["radiant_team"] = test_set["radiant_team"].str.split("-")
test_set["dire_team"] = test_set["dire_team"].str.split("-")
```

```python
def hero_pairs(team_composition):
    list_of_hero_pair = []
    for hero1 in team_composition[0]:
        for hero2 in team_composition[1]:
            list_of_hero_pair.append((int(hero1),int(hero2)))
    return list_of_hero_pair
```

```python
def winrate_pairs(hero_pairs, winrate_dataframe):
    pair_winrates = []
    winrate = 0
    for pair in hero_pairs:
        winrate = winrate_dataframe.loc[(winrate_dataframe["hero"] == pair[0]) & (winrate_dataframe["opponent"] == pair[1]), "win_rate"]
        print(pair,winrate)
        pair_winrates.append(winrate)
    return pair_winrates
```

```python
def combined_winrate(winrate_pairs):
    ### Make sure to return a winrate, even if the pair is returned as (number,NaN)
    combined_winrate = np.mean(winrate_pairs)
    return combined_winrate
```

```python
def winrate_for_team(test_dataframe, winrate_dataframe):
    radiant_comp = test_dataframe["radiant_team"]
    dire_comp = test_dataframe["dire_team"]
    team_winrates = []
    for i in range(len(radiant_comp)):
        team_composition = [radiant_comp[i],dire_comp[i]]
        ### first we compute the hero pairs
        hero_pair = hero_pairs(team_composition)
        ### second we generate the winrate between each hero on radiant and dire
        winrate_pair = winrate_pairs(hero_pair,winrate_dataframe)
        ### third we combine the winrate into a team winrate
        team_winrate = combined_winrate(winrate_pair)
        team_winrates.append(combined_winrate(winrate_pair))
        # if team_winrate > 0.5:
        #     print(f"Radiant has the advantage and a chance of {round(team_winrate * 100,1)} to win")
        # elif team_winrate <= 0.5:
        #     print(f"Dire has the advantage and a chance of {round((1-team_winrate) * 100,1)} to win")
        # elif team_winrate == None:
        #     print("Not enough hero data to tell a winrate")
    return team_winrates
```
