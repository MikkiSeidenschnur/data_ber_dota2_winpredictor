**4. Start filtering the data into MMR brackets**

```python
brackets = [4000, 4500, 5000, 5500, 6000]

def bracket_separator(dataframe,brackets):
    dota_matches_dataframes_list = []
    for bracket in brackets:
        data = dataframe.loc[(dataframe["avg_mmr"] >= bracket) & (dataframe["avg_mmr"] < bracket+500)]
        dota_matches_dataframes_list.append(data)
    return dota_matches_dataframes_list
```

```python
a, b, c, d, e = bracket_separator(dota_matches,brackets)
```
