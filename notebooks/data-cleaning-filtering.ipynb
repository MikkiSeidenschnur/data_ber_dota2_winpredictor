### We have to make sure that there are no duplicate rows, since the API returns 100 random matches. We will drop duplicates

**1. Start by importing the csv file**

```python
#dota_matches = pd.read_csv("~/Dropbox/Dotadata/")
dota_matches = pd.read_csv("../rawdata/1593961990.csv")
dota_matches
```

**2. Once the file has imported, we need to remove (drop) duplicate rows, if there is any**

```python
dota_matches = dota_matches.drop_duplicates("match_id")
dota_matches
```

**3. Make the Â´radiant_teamÂ´ & Â´dire_teamÂ´ into lists instead of strings**

```python

def make_string_list_team_combos(dataframe):
    if len(dataframe["radiant_team"][0]) != 5:
        radiant_team = []
        dire_team = []
        new_dataframe = dataframe
        for cell in new_dataframe["radiant_team"]:
            data = cell.replace("-", ",")
            data = data.split(",")
            radiant_team.append(data)
        for cell in new_dataframe["dire_team"]:
            data = cell.replace("-", ",")
            data = data.split(",")
            dire_team.append(data)
        new_dataframe["radiant_team"] = radiant_team
        new_dataframe["dire_team"] = dire_team
        return new_dataframe
    else:
        return dataframe
    
dota_matches = make_string_list_team_combos(dota_matches)
dota_matches
```

**4. Start filtering the data into MMR brackets**

```python
brackets = [4000, 4500, 5000, 5500, 6000]

def bracket_separator(dataframe,brackets):
    dota_matches_dataframes_list = []
    for bracket in brackets:
        data = dataframe.loc[(dataframe["avg_mmr"] >= bracket) & (dataframe["avg_mmr"] < bracket+500)]
        dota_matches_dataframes_list.append(data)
    return dota_matches_dataframes_list
```

```python
a, b, c, d, e = bracket_separator(dota_matches,brackets)
```
